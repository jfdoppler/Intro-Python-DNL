```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def campo_vector(z, t):
    x = z[0]
    y = z[1]
    dxdt = x+4*y
    dydt = y
    return [dxdt, dydt] 

dt = 0.001
t = np.arange(0, 30, dt)
xi = -100*np.ones(10)
yi = np.linspace(0.9,1,10)
#zi = [xi, yi]
xt=[]
yt=[]
for i in range(10):
    sol = odeint(campo_vector, [xi[i], yi[i]], t)
    xt.append(sol[:, 0])
    yt.append(sol[:, 1])
    plt.plot(xt[i], yt[i])
plt.xlim(-1.6*10**12,0.6*10**12)
plt.ylim(0,0.14*10**13)
```

# Ej Punto Saddle

```python

def campo_vector(z, t):
    A=np.matrix([[5,-2],[-17,-5]])
    x = z[0]
    y = z[1]
    dxdt = A[0,0]*x+A[0,1]*y
    dydt = A[1,0]*x+A[1,1]*y
    return [dxdt, dydt]   
```

```python
A=np.matrix([[5,-2],[-17,-5]])
def campo_vector(z, t):
    x = z[0]
    y = z[1]
    dxdt = A[0,0]*x+A[0,1]*y
    dydt = A[1,0]*x+A[1,1]*y
    return [dxdt, dydt] 
dt = 0.001
t = np.arange(0, 0.4,dt)
xi = 1.2
yi = 9
zi = [xi, yi]
sol = odeint(campo_vector, zi, t)
xt = sol[:, 0]
yt = sol[:, 1]
plt.plot(xt, yt,'b')
plt.arrow(xt[-2],yt[-2],np.diff(xt)[-1],np.diff(yt)[-1],head_width=0.5)
eig,vec=np.linalg.eig(A)
#type(eig[0]),type(eig[1])
plt.plot([-7*vec[0,0],7*vec[0,0]],[-7*vec[1,0],7*vec[1,0]],'r')
plt.plot([-7*vec[0,1],7*vec[0,1]],[-7*vec[1,1],7*vec[1,1]],'r')
```
